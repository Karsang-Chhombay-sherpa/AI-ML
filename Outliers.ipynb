```python
import numpy as np
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
q1=np.percentile(arr,25)
q3=np.percentile(arr,75)
q3,q1
```

```python
iqr=q3-q1
iqr
```

```python
UF=q3+1.5*(iqr)
LF=q1-1.5*(iqr)
UF,LF
```

```python
l=[]
for i in arr:
  if i<=UF and i>=LF:
    l.append(i)
arr2=np.array(l)
arr2
```

```python
import seaborn as sns
sns.boxplot(x=arr2)
```
