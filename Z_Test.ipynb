```python
import numpy as np
from scipy.stats import norm
```

```python
sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]
population_mean=170
population_std=3
samp_mean=np.mean(sample)
n=len(sample)
```

```python
z_score=float((samp_mean-population_mean)/(population_std/np.sqrt(n)))
z_score
```

```python
p_value=2*(1-norm.cdf(abs(z_score)))
float(p_value)
```

```python
alpha=0.05
if p_value<alpha:
  print("reject null hypothesis")
else:
  print("accept the null hypothesis")
```
